```python
#!/usr/bin/python
import pgdb
import pg

# Creamos la conexión
conexion = pg.DB(host="crzuxnovk7a2.us-east-1.rds.amazonaws.com", dbname="postgres", user="ds_test", passwd="")

# Ejecutamos una consulta y guaramos los resultados
resultados = conexion.query("SELECT id FROM orders")
Ids = []
# Recorremos los resultados y los mostramos
for id in resultados.getresult():
    Ids.append(int(id[0]))
    #print(int(id[0]))

# Cerramos la conexión
conexion.close()
```

```python
conexion = pg.DB(host="crzuxnovk7a2.us-east-1.rds.amazonaws.com", dbname="postgres", user="ds_test", passwd="")

Content = []

for id in Ids:
    Content.append(conexion.query("SELECT Pr.pos_id, e.created_at - o.order_date as difff \
    FROM events E \
    INNER JOIN products Pr ON E.order_id=Pr.Order_id \
    INNER JOIN orders o on o.id = pr.order_id \
    WHERE E.order_id='{}' AND status='hand_to_domiciliary'".format(id)).getresult())

    if len(Content) > 1000: break

conexion.close()
```

```python
import pandas as pd
import math
```

```python
Productos = []
Tiempo = []
for a in range(len(Content)):
    Parcial_P = []
    #Parcial_T = []
    for i in range(len(Content[a])):
        #Parcial_P.append(np.float(Content[a][i][0]))
        Parcial_P.append(pd.to_numeric(Content[a][i][0],errors='coerce'))
        Var = round(Content[a][i][1].seconds/60, 2)
        #if len(Parcial_T) == 0: Parcial_T.append(round(Content[a][i][1].seconds/60, 2))
        #print(Content[a][i][1])
    Productos.append(Parcial_P)
    Tiempo.append(Var)
```

```python
print(len(Productos))
print(len(Tiempo))
```

```python
Productos_F = []
Tiempo_F = []

for i, a in zip(Productos, Tiempo):
    if i == []:
        pass
    else:
        Productos_F.append(i)
        Tiempo_F.append(a)
```
